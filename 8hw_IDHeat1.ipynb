```python
#homework 8
#ashley woodward 

#Tintial lava= 1230 C
#T surrounding rock = 20 C
#dike width (d) = 5m, indef long in y and z directions 
#heat capacity 1.3 Joule/gramm/kelvin
#density (p) = 2.75 gramm/cm^3
#heat conductance  0.052 J/cm/s/kelvin

#dike N=50, dicritize 
#host rock d*2 on each side 
#total number of intervals = 5*N
#dirichlet boundary conditions T=20 on each side 
```

```python
#load libraries 
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 50*5
d = 5
L = 5*d
T = np.linspace(0,L,N+1)
x = np.linspace(0,L,N+1)


for i in range (N+1):
    if x[i] < 10 or x[i] > 15:
        T[i] = 20 + 273 #0 C = 273K
    if x[i] <= 15 and x[i] >= 10:
        T[i] = 1230 + 273 #temp in K


plt.ylim(0, 1700)
plt.plot(x,T, 'ro-',mfc='w')
plt.show() 
```

```python
#quesiton 1

#UNITS

#length unit = m
#time unit = s
#coefficient k = 0.052 J/cm/s/K = 5.2 J/m/s/K
#grid spacing (delta x) = (L/N)^2 m^2
#time step (delta t) = 3600 s (1hr)
#coeff n = (delta_t/delta_x^2)*(k/(c*p)) 
#heat capacity (c) = 1.3 J/g/K = 1300 J/kg/K
# density (p) = 2.75 g/cm^3 = 2750 kg/m^3

k = 5.2 #J/m/s/K
c = 1300 #J/kg/K
p = 2750 #kg/m^3
delta_t = 3600//(5**2) #sec
delta_x = (L/N)**2
n = (delta_t/delta_x)*(k/(c*p)) 

print("n:", n)

#Main question of this problem: What is the temperature in the middle of the dike after 
#(a) one hour, (b) one day, (c) one week, (d) one month, and (e) after one year?
```

```python

#how temp changes 

Tnew = np.copy(T)
for i in range(N):
    for j in range(N):
        Tnew[i] = n*(T[i+1]-2*T[i]+T[i-1])+T[i]
    T = Tnew #copy back to T 
    
print(T)


```

```python

#step is 1 hr = 3600s

delta_t = 3600 #sec
hour = 60*60 
day = hour*24
week = day*7
month = day*30
year = day*365


#1 HOUR 

for t in range ((hour//delta_t)):
    Tnew = np.copy(T)
    for i in range(N-1):
        for j in range(N-1):
            Tnew[i] = n*(T[i+1]-2*T[i]+T[i-1])+T[i]
        T = Tnew

print("1 hour temp at center:", max(T))
plt.plot(x,T, 'ro-',mfc='w')
plt.ylim(0,1700) 
plt.xlim(0,25)
plt.show()

        
        
```

```python
  
#1 DAY    

for t in range ((day//delta_t)):
    Tnew = np.copy(T)
    for i in range(N-1):
        for j in range(N-1):
            Tnew[i] = n*(T[i+1]-2*T[i]+T[i-1])+T[i]
        T = Tnew

print("1 day temp at center:", max(T))
plt.plot(x,T, 'ro-',mfc='w')
plt.ylim(0,1700) 
plt.xlim(0,25)
plt.show()
     
```

```python
#1 Week   

for t in range ((week//delta_t)):
    Tnew = np.copy(T)
    for i in range(N-1):
        for j in range(N-1):
            Tnew[i] = n*(T[i+1]-2*T[i]+T[i-1])+T[i]
        T = Tnew

print("1 day temp at center:", max(T))
plt.plot(x,T, 'ro-',mfc='w')
plt.ylim(0,1700) 
plt.xlim(0,25)
plt.show()
     
```

```python

#1 MONTH

for t in range ((30*24*60*60//1800)):
    Tnew = np.copy(T)
    for i in range(N-1):
        Tnew[i] = n*(T[i+1]-2*T[i]+T[i-1])+T[i]
    T = Tnew

print("1 month temp at center:", max(T))
plt.plot(x,T, 'ro-',mfc='w')
plt.ylim(0,1700) 
plt.xlim(0,25)
plt.show()
```

```python

#1 YEAR

for t in range ((year//delta_t)):
    Tnew = np.copy(T)
    for i in range(N-1):
        Tnew[i] = n*(T[i+1]-2*T[i]+T[i-1])+T[i]
    T = Tnew

print("1 year temp at center:", max(T))
plt.plot(x,T, 'ro-',mfc='w')
plt.ylim(0,1700) 
plt.xlim(0,25)
plt.show()
```
