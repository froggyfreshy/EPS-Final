```python
#question 2

N = 20
L = 1

x = np.linspace(0,L,N+1)
T = np.linspace(0,L,N+1)

for i in range (N+1):
    if x[i] <= 0.5:
        T[i] = 2*x[i]
    if x[i] > 0.5:
        T[i] = 2-2*x[i]

            
plt.ylim(0,1)       
plt.plot(x,T, 'ro-',mfc='w')
plt.show()

```

```python
#question 3

K = 1
dt = 10e-4
dx = 0.05
k = 1

        
constant = (k*dt)/(dx**2)
print(constant)

Tnew = np.copy(T)
for i in range(N):
    for j in range(N):
        '''
        T[0] = 0  
        T[1] = 0
        '''
        Tnew[i] = constant*(T[i+1]-2*T[i]+T[i-1])+T[i]
    T = Tnew #copy back to T 
    
print(T)


```

```python
#question 4 and 5 plot temp dist at each step 

for _ in range (240):
    Tnew = np.copy(T)
    for i in range(N):
        for j in range(N):
            Tnew[i] = constant*(T[i+1]-2*T[i]+T[i-1])+T[i]
        T = Tnew
    if _%20 == 0:
        print("percent of intial temp (%):",max(T)*100)
        plt.clf()
        plt.plot(x,T, 'ro-',mfc='w')
        plt.ylim(0,1) 
        plt.xlim(0,1) 
        plt.show()
        plt.draw()
        plt.pause(0.05)
        
```
