```python
#QUESTION 1
k = 0.00012097
t = 0.0
c = 1.0
```

```python
tt = [t]
cc = [c]

dt = 10
tmax = 10000

while (t<tmax):
    c = c*(1-k*dt)
    t = t+dt
    tt.append(t)
    cc.append(c)
```

```python
#QUESTION 2
ms = 3e+6
mp = 1
G  = 1e-6
```

```python
t = 0
tmax = 100
dt   = 10e-3
n    = 10/dt
r = np.array([1,0])
v = np.array([0,2])
```

```python
#for loop calc time derivative 
vx = 0
vy = 2
x = 1
y = 0 

xt = []
yt = [] 

while (t <tmax):
    r = np.sqrt(x**2+y**2)
    dvx = -(G*ms/r**3)*x*dt
    dvy = -(G*ms/r**3)*y*dt
    vx = vx+dvx
    vy = vy+dvy
    dx = (vx)*dt
    dy = (vy)*dt
    x = x+dx
    y = y+dy
    t = t+dt 
    xt.append(x)
    yt.append(y)

```

```python
plt.plot(xt,yt,'r-')
plt.plot(0,0,'*',mfc='w',ms=10)
plt.xlim(-3,3)
plt.ylim(-3,3)

plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
#QUESTION 3
r = np.array([1,0])
v = np.array([0,2])



def KeplerODE(t,y):
    global mp,ms,G
    r = y[0:2]
    v = y[2:4]
    
    drdt = v
    
    F = -ms*mp*G/np.linalg.norm(r)**3 * r
    a = F/mp 
    dvdt = a 
    
    return np.concatenate((drdt,dvdt))
```

```python
t = 0
tmax = 100
dt   = 10e-3


tt = []
xt = []
yt = [] 

y = np.concatenate((r,v))
while (t<tmax):
    r = y[0:2]
    v = y[2:4]
    tt.append(t)
    xt.append(r[0])
    yt.append(r[1])
    
    dydt = KeplerODE(t,y)
    y = y+dydt*dt
    t = t +dt 
    
```

```python
plt.plot(xt,yt,'r-')
plt.plot(0,0,'*',mfc='w',ms=10)
#plt.xlim(-3,3)
#plt.ylim(-3,3)

plt.gca().set_aspect('equal', adjustable='box')
plt.show()

#revisit and correct this problem 
```
