```python
#Part 2 2D Heat Eqs 

N = 51+1
L = 51+1

x = np.linspace(0,L,N+1)
y = np.linspace(0,L,N+1)
T = np.zeros((L,L))

for i in range(N):
    T[i][0] = 100
    
for _ in range(2000):
    Tnew = np.copy(T)
    for i in range(1,N-1):
        for j in range(1,N-1):
            dxm = T[i-1][j]
            dxp = T[i+1][j]
            dym = T[i][j-1]
            dyp = T[i][j+1]
            Tnew[i][j] = (dxp + dxm + dyp + dym)/4
    T = Tnew
            
```

```python
#run in 2D

n=51+1
L=51+1

from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
X = np.linspace(0, L, n) # where n is number of segments
Y = np.linspace(0, L, n)
X, Y = np.meshgrid(X, Y)
fig = plt.figure(dpi=200)

for i in range(n):
    T[i][0] = 100
    
for _ in range(2000):
    Tnew = np.copy(T)
    for i in range(1,n-1):
        for j in range(1,n-1):
            dxm = T[i-1][j]
            dxp = T[i+1][j]
            dym = T[i][j-1]
            dyp = T[i][j+1]
            Tnew[i][j] = (dxp + dxm + dyp + dym)/4
    fig.clear()
    ax = fig.gca(projection='3d')
    ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, antialiased=False)
    plt.draw()
    plt.pause(0.05)
```
